## Question 1:
### Write a function that given your birthday (as a date), returns how old you are in years.

```python
import datetime

def age(d, m, y):
  year = datetime.datetime.now().year
  r = year - y
  return r

print("Enter day, month, year")    # Eg: 02 04 2000
d, m, y = map(int, input().split())
result = age(d, m, y)
print("Age: ", result)
```

## Question 6:
### Were operated by United, American, or Delta

```python
dataset[(dataset['carrier'] == 'AA') | (dataset['carrier'] == 'UA') | (dataset['carrier'] == 'DL')] 
```

## Question 7:
### Departed in summer (July, August, and September)

```python
dataset[(dataset['month'] > 6) & (dataset['month'] < 10)]
```

## Question 8:
### Arrived more than two hours late, but didn???t leave late

```python
dataset[(dataset['arr_delay'] > 120) & (dataset['dep_delay']  <= 0)]
```

## Question 9:
### Were delayed by at least an hour, but made up over 30 minutes in flight

```python
dataset[(dataset['dep_delay'] > 60) & (dataset['air_time'] > 30)]
```
